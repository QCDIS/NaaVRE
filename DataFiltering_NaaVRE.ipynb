```r
library(dplyr)

#datain is the input file 
#clusterin one or more among country,locality,year,month,day,parenteventid,eventid
#taxlev one among scientificname,kingdom,phylum,class,order,family,genus
#paramin one or more among totalbiovolume,totalcarboncontent,density
#thresholdin between 0 and 1

param_datain <- '~/Unisalento/Lifewatch/Phyto_VRE/Input/TraitsOutput_Advanced_Progetto_Strategico_Australia.csv'

param_cluster_whole <- 0
param_cluster_country <- 1
param_cluster_locality <- 1
param_cluster_year <- 1
param_cluster_month <- 1
param_cluster_day <- 1
param_cluster_parenteventid <- 1
param_cluster_eventid <- 1

param_taxlev <- 'scientificname'

param_totalbiovolume <- 1
param_totalcarboncontent <- 0
param_density <- 0

param_threshold <- 0.75

dataset <- read.csv(param_datain,stringsAsFactors=F,sep = ";", dec = ".")

clusterin <- c()
if (param_cluster_country==1) clusterin=append(clusterin,"country")
if (param_cluster_locality==1) clusterin=append(clusterin,"locality")
if (param_cluster_year==1) clusterin=append(clusterin,"year")
if (param_cluster_month==1) clusterin=append(clusterin,"month")
if (param_cluster_day==1) clusterin=append(clusterin,"day")
if (param_cluster_parenteventid==1) clusterin=append(clusterin,"parenteventid")
if (param_cluster_eventid==1) clusterin=append(clusterin,"eventid")


if(!'density'%in%names(dataset))dataset["density"]=1
if(!'biovolume'%in%names(dataset))dataset$biovolume=1
if(!'cellcarboncontent'%in%names(dataset))dataset$cellcarboncontent=1
if(!'totalbiovolume'%in%names(dataset))dataset$totalbiovolume=dataset$biovolume*dataset["density"]
if(!'totalcarboncontent'%in%names(dataset))dataset$totalcarboncontent=dataset$cellcarboncontent*dataset["density"]

write.table(dataset,paste('~/Unisalento/Lifewatch/Phyto_VRE/Script_R/Data_Filtering/dfdataset.csv',sep=''),row.names=F,sep = ";",dec = ".",quote=F)
```

```r
dataset=read.csv('~/Unisalento/Lifewatch/Phyto_VRE/Script_R/Data_Filtering/dfdataset.csv',stringsAsFactors=F,sep = ";", dec = ".")
ClusterWhole=param_cluster_whole
cluster=clusterin
taxlev=param_taxlev
TraitDensity=param_density

if(ClusterWhole==0) {
  if(length(cluster)>1) ID<-apply(dataset[,cluster],1,paste,collapse='.')
  if(length(cluster)==1) ID<-dataset[,cluster]
} else if(ClusterWhole==1) {
  ID<-rep('all',dim(dataset)[1]) }

if (TraitDensity==1) {
  
  IDZ<-unique(ID)  
  IDLIST<-list()
  length(IDLIST)<-length(IDZ)
  names(IDLIST)<-IDZ
  
  # ranked distribution of the taxa
  for(j in 1:length(IDZ)){
    ddd<-dataset[ID==IDZ[j],]
    totz<-sum(ddd[,'density'],na.rm=T)
    matz<-tapply(ddd[,'density'],ddd[,taxlev],function(x)sum(x,na.rm=T)/totz)
    matz<-sort(matz,decreasing=T) 
    
    # cumulative contribution to the overall density
    k<-2
    trs<-max(matz)
    while (trs<threshold) {
      matz[k]<-matz[k-1]+matz[k]
      trs<-matz[k]
      k<-k+1 }
    
    matzx<-matz[1:k-1]
    
    IDLIST[[j]] <- ddd[ddd[,taxlev]%in%names(matzx),]
  }
  
  # filtered dataset for density
  dataset.d <- do.call('rbind',IDLIST)
  
} else dataset.d <- dataset[FALSE,]

write.table(dataset.d,paste('~/Unisalento/Lifewatch/Phyto_VRE/Script_R/Data_Filtering/datasetD.csv',sep=''),row.names=F,sep = ";",dec = ".",quote=F)
```
